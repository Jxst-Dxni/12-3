```python
import math
import matplotlib.pyplot as plt
import numpy as np


# Konstanten
g = 9.81            
v = 30.0            
dt = 0.01           
winkel = 60         


vx = v * math.cos(math.radians(winkel))
vy = v * math.sin(math.radians(winkel))


x = 0
y = 0


xR = [x]
yR = [y]


while y >= 0:
    vy += -g * dt     
    x += vx * dt      
    y += vy * dt      
    
    xR.append(x)
    yR.append(y)

    if y < 0:
        print(vx)
        print(vy)
        break
plt.plot(xR, yR)
plt.title("Wurfparabel")
plt.xlabel("Entfernung (m)")
plt.ylabel("Höhe (m)")
plt.ylim(bottom=0)
plt.grid(True)
plt.show()
```
