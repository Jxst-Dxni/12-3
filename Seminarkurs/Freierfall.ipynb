```python
import math
import numpy as up

x = up.arange(1,11)
print(x)
y = x**2
print(y)
from matplotlib import pyplot as plt 
plt.plot(x,y)
plt.show()
```

```python
x = up.arange(-10, 10.1, 0.1)
y = x**2
plt.plot(x, y)
plt.xlim(-10, 10)
plt.ylim(-10, 110)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
plt.plot(x, y, color="red")
```
