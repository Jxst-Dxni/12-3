```python
import pygame
import random
import math

# -------------------------------
# Einstellungen der Simulation
# -------------------------------

WIDTH = 800      # Breite des Fensters
HEIGHT = 600     # Höhe des Fensters
NUM_BALLS = 10   # Anzahl der Kugeln
RADIUS = 10      # Radius jeder Kugel

pygame.init()
screen = pygame.display.set_mode((WIDTH, HEIGHT))
clock = pygame.time.Clock()

# ---------------------------------
# Klasse für eine Kugel
# ---------------------------------

class Ball:
    def __init__(self):
        # zufällige Startposition
        self.x = random.randint(RADIUS, WIDTH - RADIUS)
        self.y = random.randint(RADIUS, HEIGHT - RADIUS)

        # zufällige Geschwindigkeit
        self.vx = random.uniform(-3, 3)
        self.vy = random.uniform(-3, 3)

    def move(self):
        """
        Bewegung der Kugel:
        neue Position = alte Position + Geschwindigkeit
        """
        self.x += self.vx
        self.y += self.vy

    def wall_collision(self):
        """
        Prüft Kollision mit Wänden.
        Wenn Kugel die Wand berührt -> Geschwindigkeit umkehren
        """
        if self.x <= RADIUS or self.x >= WIDTH - RADIUS:
            self.vx *= -1

        if self.y <= RADIUS or self.y >= HEIGHT - RADIUS:
            self.vy *= -1

    def draw(self):
        """Zeichnet die Kugel"""
        pygame.draw.circle(screen, (255, 255, 255), (int(self.x), int(self.y)), RADIUS)


# ---------------------------------
# Kugel-Kollision
# ---------------------------------

def ball_collision(ball1, ball2):
    """
    Prüft ob zwei Kugeln zusammenstoßen.
    Wenn ja -> elastischer Stoß
    """

    dx = ball1.x - ball2.x
    dy = ball1.y - ball2.y

    distance = math.sqrt(dx**2 + dy**2)

    # Wenn Abstand kleiner als 2 Radien -> Kollision
    if distance < 2 * RADIUS:

        # einfache elastische Stoßregel:
        # Geschwindigkeiten tauschen
        ball1.vx, ball2.vx = ball2.vx, ball1.vx
        ball1.vy, ball2.vy = ball2.vy, ball1.vy


# ---------------------------------
# Kugeln erzeugen
# ---------------------------------

balls = [Ball() for _ in range(NUM_BALLS)]

# ---------------------------------
# Hauptschleife der Simulation
# ---------------------------------

running = True
while running:

    clock.tick(60)  # 60 Zeitschritte pro Sekunde

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    screen.fill((0, 0, 0))  # Hintergrund schwarz

    # Bewegung und Wandkollision
    for ball in balls:
        ball.move()
        ball.wall_collision()

    # Kugel-Kollisionen prüfen
    for i in range(NUM_BALLS):
        for j in range(i + 1, NUM_BALLS):
            ball_collision(balls[i], balls[j])

    # Kugeln zeichnen
    for ball in balls:
        ball.draw()

    pygame.display.flip()

pygame.quit()
```
