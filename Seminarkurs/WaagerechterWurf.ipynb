```python
# ...existing code...
import math
import numpy as np
import matplotlib.pyplot as plt

g = 9.81
v = 30.0
dt = 0.01
winkel = 60

vx = v * math.cos(math.radians(winkel))
vy = v * math.sin(math.radians(winkel))

xR = []
yR = []

for i in range(10000):  # genug Schritte, wird durch Abbruch bei y<0 gestoppt
    t = i * dt
    x = vx * t
    y = vy * t - 0.5 * g * t**2

    if y < 0:
        print(vx)
        print(vy)
        break

    xR.append(x)
    yR.append(y)

plt.plot(xR, yR)
plt.title("Projectile Motion")
plt.xlabel("Distance (m)")
plt.ylabel("Height (m)")
plt.ylim(bottom=0)
plt.grid()
plt.show()
# ...existing code...
```
