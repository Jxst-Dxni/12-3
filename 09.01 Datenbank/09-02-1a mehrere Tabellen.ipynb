#### Aufgabe 1: INNER JOIN

```python
SELECT
    schueler.vorname,
    schueler.nachname,
    projekt.titel,
    projekt.fachbereich
FROM schueler
INNER JOIN projekt
ON schueler.projekt_id = projekt.projekt_id;
```

#### Beobachtung
Noah fehlt, weil seine projekt_id den Wert NULL hat. Die Wetterstation fehlt, weil kein Schüler diesem Projekt zugeordnet ist.

| Verbindungsart | Zweck |
| --- | --- |
| INNER JOIN | Zeigt nur Zeilen, bei denen der Wert aus schueler.projekt_id zu einem Wert aus projekt.projekt_id passt. |

#### Aufgabe 2: LEFT JOIN


```python
SELECT
    schueler.vorname,
    schueler.nachname,
    projekt.titel,
    projekt.fachbereich
FROM schueler
LEFT JOIN projekt
ON schueler.projekt_id = projekt.projekt_id;
```

#### Beobachtung

| Verbindungsart | Zweck |
| --- | --- |  

Zeigt alle Zeilen der linken Tabelle und ergänzt passende Daten aus der rechten Tabelle

#### Aufgabe 3: RIGHT JOIN

```python
SELECT
    schueler.vorname,
    schueler.nachname,
    projekt.titel,
    projekt.fachbereich
FROM schueler
RIGHT JOIN projekt
ON schueler.projekt_id = projekt.projekt_id;
```

#### Beobachtung


| Verbindungsart | Zweck |  

Zeigt alle Zeilen der rechten Tabelle und ergänzt passende Daten aus der linken Tabelle


#### Aufgabe 4: FULL OUTER JOIN


```python
SELECT
    schueler.vorname,
    schueler.nachname,
    projekt.titel,
    projekt.fachbereich
FROM schueler
FULL OUTER JOIN projekt
ON schueler.projekt_id = projekt.projekt_id;
```

#### Beobachtung


| Verbindungsart | Zweck |
| --- | --- |  

Es wird alles gelistet und wenn Informationen fehlen werden sie durch den Wert Null in der Zeile ergänzt.


#### Aufgabe 5: Verbindung mit WHERE ohne JOIN

```python
SELECT
    schueler.vorname,
    schueler.nachname,
    projekt.titel,
    projekt.fachbereich
FROM schueler, projekt
WHERE schueler.projekt_id = projekt.projekt_id;
```

#### Beobachtung


| Verbindungsart | Zweck |
| --- | --- |  
erbindet  zwei Tabellen über eine Bedingung in der WHERE-Klausel.  Entspricht hier einem Inneren Join.
