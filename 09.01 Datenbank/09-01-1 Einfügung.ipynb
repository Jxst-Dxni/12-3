3.2 **Beschreibe** die Bedeutung von:
    - 
 - Fremdschlüssel:  
    - Verknüpfung: Er ist eine Spalte in einer Tabelle, die auf den Primärschlüssel einer anderen Tabelle verweist.  
    - Beziehung: Er schafft eine logische Verbindung (z. B. steht in der Tabelle „Bestellungen“ die „Kundennummer“ als Fremdschlüssel, um die Bestellung einem Kunden aus der Tabelle „Kunden“ zuzuordnen).  
    - Integrität: Er sorgt dafür, dass keine Verweise auf Datensätze existieren, die es gar nicht gibt.   
- Primärschlüssel: 
    - Eindeutigkeit: Jeder Wert darf nur ein einziges Mal in der Spalte vorkommen (z. B. eine Kundennummer).  
    - Keine Leerstellen: Er darf niemals leer sein (NOT NULL).  
    - Zweck: Er stellt sicher, dass man jede Zeile in einer Tabelle präzise identifizieren und ansprechen kann.
